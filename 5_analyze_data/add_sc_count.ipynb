```python
import pathlib
import pandas as pd
import seaborn as sns
from matplotlib.colors import rgb2hex
from mpl_toolkits.mplot3d import axes3d, Axes3D

import analysisutils as utils
```

```python
norm_data_path = pathlib.Path("../4_processing_features/data/nf1_sc_norm_fs_cellprofiler.csv.gz")
norm_data = pd.read_csv(norm_data_path, compression="gzip", index_col=0)
norm_data = norm_data.reset_index()
norm_data
```

```python
norm_data.Metadata_WellRow.value_counts()
```

```python
grouped_data = norm_data.groupby(["Metadata_WellRow", "Metadata_WellCol"])['Metadata_WellRow'].count().reset_index(name='Metadata_number_of_singlecells')
grouped_data
```

```python
number_sc = []
for row in norm_data['Metadata_Well']:
    if row == 'C6' :    number_sc.append(12)
    elif row == 'C7':   number_sc.append(12)
    elif row == 'D6':   number_sc.append(5)
    elif row == 'D7':   number_sc.append(14)
    elif row == 'E6':   number_sc.append(9)
    elif row == 'E7':   number_sc.append(44)
    elif row == 'F6':   number_sc.append(7)
    elif row == 'F7':   number_sc.append(46)
    
```

```python
# norm_data['Metadata_number_'] = rating
#     print(df) 

norm_data.insert(2, "Metadata_number_of_singlecells", number_sc)
norm_data
```

```python
output_dir = "data"
norm_feature_select_output = pathlib.Path(f"{output_dir}/nf1_sc_norm_fs_cellprofiler_edit.csv.gz")

norm_data.to_csv(norm_feature_select_output)
```
